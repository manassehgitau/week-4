```python
import random

# positions (head: Tail)
snakes = {
    99: 78,
    95: 75,
    92: 88,
    64: 60,
    59: 39,
    52: 42,
    36: 6,
    16: 8,
}

ladders = {
    4: 14,
    9: 31,
    21: 42,
    28: 84,
    51: 67,
    72: 91,
    80: 99,
}

players = ["Player 1", "Player 2"]
positions = {player: 0 for player in players}

def roll_dice():
    return random.randint(1, 6)

def move_player(player):
    input(f"{player}, press Enter to roll the dice...")
    roll = roll_dice()
    print(f"{player} rolled a {roll}")
    current = positions[player]
    next_pos = current + roll

    if next_pos > 100:
        print("Roll exceeds 100. Try again next turn.")
        return

    print(f"{player} moves from {current} to {next_pos}")

    # Check for snake or ladder
    if next_pos in snakes:
        print(f"Oh no! A snake bites {player}. Going down from {next_pos} to {snakes[next_pos]}")
        next_pos = snakes[next_pos]
    elif next_pos in ladders:
        print(f"Yay! A ladder lifts {player} up from {next_pos} to {ladders[next_pos]}")
        next_pos = ladders[next_pos]

    positions[player] = next_pos
    print(f"{player} is now at position {positions[player]}")

# Game loop
def play_game():
    while True:
        for player in players:
            move_player(player)
            if positions[player] == 100:
                print(f"\nðŸŽ‰ {player} wins the game! ðŸŽ‰")
                return

# Run the game
if __name__ == "__main__":
    print("ðŸŽ² Welcome to Snakes and Ladders! ðŸŽ²\n")
    play_game()
```
